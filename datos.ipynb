```python
satisfaction_by_year = df.groupby('Year')['Satisfaction_Score'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=satisfaction_by_year, x='Year', y='Satisfaction_Score')  # Barras para comparar años
plt.title('Puntajes de Satisfacción de Estudiantes por Año')
plt.xlabel('Año')
plt.ylabel('Puntaje de Satisfacción')
plt.show()
```
